## Concat
* Similar cu concat de la numpy

```python
df1 = pd.DataFrame([['a', 1], ['b', 2]],
                   columns=['letter', 'number'])
df2 = pd.DataFrame([['c', 3, 'cat'], ['d', 4, 'dog']],
                   columns=['letter', 'number', 'animal'])
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], join="inner")
```

## Merge

```python
df1 = pd.DataFrame([['cust1', 1], ['cust2', 2]],
                   columns=['customer', 'childrens'])
df2 = pd.DataFrame([['cust1', 30, 'cat'], ['cust2', 44, 'dog']],
                   columns=['customer', 'age', 'pet'])
```

```python
df1.merge(df2, left_on='customer', right_on='customer')
```

```python
df1.merge(df2, on='customer')
```

```python
df1.merge(df2, on='customer', how='inner')
```

```python
df2 = pd.DataFrame([['cust1', 30, 'cat'], ['cust2', 44, 'dog'], ['cust3', 35, 'parrot']],
                   columns=['customer', 'age', 'pet'])
```

```python
df2.merge(df1, on='customer', how='left')
```
